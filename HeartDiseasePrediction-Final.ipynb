### Data Analysis

After exploring the data and extracting some basic information, we can begin to prepare our data for our model. Firstly, before we begin, let's split our data into training and testing set. Remember that the purpose of this analysis is to classify patients with and without heart disease using the features, `X`, so that is our target `y`. 

We can do this using the `train_test_split` method in the `sklearn` package. We selected 25% of our data to be used for training and 75% to be used for testing. 

```python
!pip install --upgrade pip
!pip install -U scikit-learn==1.1.1

!pip install numpy 
!pip install scipy
#from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split

features = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope']
X = heart[features]
y = heart["HeartDisease"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size = .75)
```

Now, we can begin to transform the data. Since this is a classification problem, we should begin with the LogisticRegression model, whose implementation in the `sklearn` package requires all data to be numerical. 

For our categorical columns, we can use encoding to convert the data to numerical. In this specific case, all of our columns do not have values that are ordinally labeled or that can be split into levels of increasing or decreasing value. Therefore, we can use one-hot encoding for each categorical column. 

For our numerical columns, we can apply scaling because their corresponding distributions have some skews, as identified during the data exploration phase. Here, we used standard scaling, we subtract the mean of the distribution and scale by unit variance. This results in values that are all around the same range.

Below, we have created two transformers; one that uses `OneHotEncoder` on the categorical columns and one that uses `StandardScaler` on the remaining numerical columns.

We then apply the transformers separately on the training data and print some information about the transformed training data set. 

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

transformer1 = ColumnTransformer([
    ('ohe', OneHotEncoder(), ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'])], 
    remainder='passthrough')

transformer2 = ColumnTransformer([
    ('ss', StandardScaler(), ['remainder__Age', 'remainder__RestingBP', 'remainder__Cholesterol', 'remainder__FastingBS', 'remainder__MaxHR', 'remainder__Oldpeak'])],
    remainder='passthrough')


transformed1 = transformer1.fit_transform(X_train)
X_train_transformed = pd.DataFrame(transformed1, columns=transformer1.get_feature_names_out())

transformed2 = transformer2.fit_transform(X_train_transformed)
X_train_transformed2 = pd.DataFrame(transformed2, columns=transformer2.get_feature_names_out())

X_train_transformed2.columns
```

Now, we can apply the same transformations to the testing data set.

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LogisticRegression

transformed1t = transformer1.fit_transform(X_test)
X_test_transformed = pd.DataFrame(transformed1t, columns=transformer1.get_feature_names_out())

transformed2t = transformer2.fit_transform(X_test_transformed)
X_test_transformed2 = pd.DataFrame(transformed2t, columns=transformer2.get_feature_names_out())
```

Finally, with our two datasets, we can first fit the regressor, `LogisticRegression`, to our transformed training data. Likewise, we can predict our y variable using testing data and and calculate our accuracy score, or the ratio of the sum of true positives and true negatives to the total number of observations. Our accuracy is around 87%, which is fair performance for our model.

```python
from sklearn.metrics import accuracy_score
regressor = LogisticRegression()

regressor.fit(X_train_transformed2, y_train)
y_pred = regressor.predict(X_test_transformed2)
accuracy_score(y_test, y_pred)
```

We can take a deeper dive into our model by printing a confusion matrix. Our model correctly predicted 87 patients who have heart disease and 113 patients who do not have heart disease. In other words, our model had 87 true positives and 113 true negatives. It appears that our model incorrectly predicted 11 people to have heart disease and 19 people not to have heart disease. There are slightly more patients with heart disease in the dataset, so perhaps it might improve the accuracy of our model if more patients without heart disease are added.

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels=regressor.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=regressor.classes_)
disp.plot()
plt.show()
```

Rather than using our `train_test_split` method, we can perform cross validation to make us more confident that our model is correctly being trained on the training data. It guarantees that our model performance of 87% is not dependent on how we split our data into train and test sets. First, we concatenate our train and test set after they were transformed. This splitting will be taken care of automatically.

```python
X_merged = pd.concat([X_train_transformed2,X_test_transformed2], ignore_index=True)
y_merged = pd.concat([y_train, y_test], ignore_index=True)
```

Here, we are using k-fold cross validation with 10 folds. This means our data set is split into 10 subsets, 9 of which compose our training set and the remainder comprises our test set. This process is run 10 times so that each subset has an opportunity to be a test set. Then, the accuracy scores of each run are averaged, which results in a k-fold cross validation score.

Our performance metric here is 0.862, or 86.2% accuracy, which is similar to our initial metric.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import std

cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model
model = LogisticRegression()
# evaluate model
scores = cross_val_score(model, X_merged, y_merged, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

While logistic regression is quite common in binary classification, a descision tree model can also be advantageous. Let's implement that model to determine its performance. Similarly, we will create the classifier, it it to our training data, predict our testing data and then compare the predicted and actual values. The accuracy score for this model is 81%, lower than that of the logistic regression model.

```python
from sklearn.tree import DecisionTreeClassifier

descisiontreeclf = DecisionTreeClassifier()

descisiontreeclf.fit(X_train_transformed2, y_train)
y_pred = descisiontreeclf.predict(X_test_transformed2)
accuracy_score(y_test, y_pred)
```

Comparing the logistic regression confusion matrix to the confusion matrix below, we can notice that there are less true positives and true negatives. So, from this basic descision tree model, it appears that logistic regression is superior. 

```python
cm = confusion_matrix(y_test, y_pred, labels=descisiontreeclf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=descisiontreeclf.classes_)
disp.plot()
plt.show()
```

The `sklearn` package allows us to easily plot the descision tree using the `plot_tree` method to understand how the tree is structured. Running this code reveals that our tree is quite vast and difficult to decipher. 

```python
from sklearn.tree import plot_tree

feature_names = ['ss__remainder__Age', 'ss__remainder__RestingBP',
       'ss__remainder__Cholesterol', 'ss__remainder__FastingBS',
       'ss__remainder__MaxHR', 'ss__remainder__Oldpeak',
       'remainder__ohe__Sex_F', 'remainder__ohe__Sex_M',
       'remainder__ohe__ChestPainType_ASY',
       'remainder__ohe__ChestPainType_ATA',
       'remainder__ohe__ChestPainType_NAP', 'remainder__ohe__ChestPainType_TA',
       'remainder__ohe__RestingECG_LVH', 'remainder__ohe__RestingECG_Normal',
       'remainder__ohe__RestingECG_ST', 'remainder__ohe__ExerciseAngina_N',
       'remainder__ohe__ExerciseAngina_Y', 'remainder__ohe__ST_Slope_Down',
       'remainder__ohe__ST_Slope_Flat', 'remainder__ohe__ST_Slope_Up']

class_names=['0', '1']
fig = plt.figure(figsize=(25,20))
full_tree = plot_tree(descisiontreeclf, feature_names=feature_names, class_names=['0', '1'], filled=True)

#plt.show()
```

The depth of our tree is 15, meaning there are 15 edges from the root node to the lowest leaf node. We can try to create a simpler tree by tuning some of the hyperparameters in a decision tree classifier. 

```python
#https://www.displayr.com/machine-learning-pruning-decision-trees/
#https://towardsdatascience.com/entropy-how-decision-trees-make-decisions-2946b9c18c8

max_depth = descisiontreeclf.get_depth()
max_depth
```

Since our tree is quite large, we can try to limit the depth of the tree using the `max_depth` hyperparameter that can take on values between 1 to 15. Similarly, we can control the `criterion` hyperparamter, which defines the way the model selects which features provide the most information gain at each level of the tree. These three possible crerion are entropy, gini and log_loss. 

To test combintations of these hyperparameters, we use `GridSearchCV`, passing the estimator, the scoring mechanism and a list containing the dictionary of parameters and their possible values.

We then fit the data, and the model performs cross-validation using all possible combinations of hyperparameters. We can access the best parameters using the `best_params_` attribute after our grid search has been fit. Our accuracy score is now 85%, which shows and improvement from the initial descision tree.

```python
from sklearn.model_selection import GridSearchCV, ParameterGrid
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer

parameters = [{'criterion': ['entropy', 'gini', 'log_loss'], 'max_depth':[max_depth for max_depth in range(1, max_depth+1)]}]

grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), scoring='accuracy', param_grid=parameters)

grid_search.fit(X_train_transformed2, y_train)

print(grid_search.best_params_)

y_pred = grid_search.predict(X_test_transformed2)

accuracy_score(y_test, y_pred)

```

For comparison, we can plot our optimized descision tree, which is much easier to read. Taking a look at the tree discloses how the model classifies, specifically what features offer the largest information gain beginning with the the ST slope.

```python
fig = plt.figure(figsize=(25,20))


full_tree = plot_tree(grid_search.best_estimator_, feature_names=feature_names, class_names=['0', '1'], filled=True)

plt.show()
```
